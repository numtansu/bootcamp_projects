## Data

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

**จงทำการ import matplotlib.pyplot มายัง plt และ ใช้คำสั่ง %matplotlib inline ถ้าหากว่าใช้งานอยู่ jupyter notebook. หากว่าไม่ได้ใช้งาน Jupyter notebook จะสามารถใช้คำสั่งใดแทน %matplotlib inline ได้ ?**

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

## Exercise 1

**จงทำตามขั้นตอนต่อไปนี้**
* **จง plot figure object ด้วยการเรียกใช้ตัวแปร fig ด้วย plt.figure()**
* **Create a figure object called fig using plt.figure()**
* **จงใช้ add_axes เพื่อเพิ่ม figure canvas ด้วย ลิสต์ที่เท่ากับ [0,0,1,1]. และให้ปรับปรุง axis ด้วยตัวแปร ax.**
* **จง Plot (x,y) บน axes และ ทำการเติม labels และ titles ลงในรูปที่ plot**

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y);
```

## Exercise 2
**จงสร้าง figure และ axes สองแกน พร้อมกับกำหนดระยะของ axes ทั้งสองอัน ax1 และ ax2 ด้วย [0,0,1,1] และ [0.2,0.5,.2,.2] ตามลำดับ**

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])
```

**จากนั้นให้ทำการ plot (x , y ) บนแกนทั้งสอง และแสดง figure**

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])

ax1.plot(x,y)
ax2.plot(x,y)
```

## Exercise 3
**ให้ทำการ plot ด้วย axes สองชุด พร้อมกับกำหนดระยะของ axes ทั้งสองนั้นด้วย ด้วย  [0,0,1,1] และ [0.2,0.5,.4,.4] ตามลำดับ**


```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])

ax1.plot(x,y)
ax2.plot(x,y)
```

**จากนั้นให้ใช้  arrays x, y และ z เพื่อทำการ plot บนแกนที่สร้างจากข้อที่แล้ว  (Notice อย่าลืมกำหนด  x - limits และ y - limits )**

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])

ax1.plot(x,z)
ax1.set_xlabel('x')
ax1.set_ylabel('z')
ax1.set_title('x_vs_z')

ax2.plot(x,y)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('Zoom')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)
```

## Exercise 4

**จงใช้คำสั่ง plt.subplots(nrows=1, ncols=2) เพื่อทำการ plot canvas ว่าง**

```python
fig, axse = plt.subplots(nrows=1, ncols=2)
```

**จากนั้นให้ทำการ plot (x,y) และ plot (x,z) บนแกน axes ของ canvas ที่ได้สร้างไว้แล้ว. และให้ใช้งานคำสั่ง linewidth and style เพื่อตกแต่งเส้นของกราฟ**

```python
fig, axse = plt.subplots(nrows=1, ncols=2)
axse[0].plot(x,y, lw=3)
axse[1].plot(x,z,lw=5, ls='--', color='red')
```

**จากนั้น ให้ทำการ resize ด้วยคำสั่ง figsize() กับการ plot แบบ subplot ด้วยคำสั่ง plt.subplots() ซึ่งสามารถ copy และ paste code จากข้อที่แล้วได้**

```python
fig, axse = plt.subplots(figsize=(8,1),nrows=1, ncols=2)

axse[0].plot(x,y,lw=3)
axse[0].set_xlabel('x')
axse[0].set_ylabel('y')

axse[1].plot(x,z, lw=5, ls='--', color='red')
axse[1].set_xlabel('x')
axse[1].set_ylabel('z')
```
