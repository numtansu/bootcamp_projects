**จงทำการ plot ให้เหมือนกับรูปตัวอย่างด้านล่างนี้**

```python
g = sns.JointGrid(x="fare", y="age", data=titanic) 
g = g.plot(plt.scatter, sns.distplot)
```

**จงทำการ plot ให้เหมือนกับรูปตัวอย่างด้านล่างนี้**

```python
plt.hist(titanic['fare'],bins=20,color='red' )
plt.xlabel('fare')
```

**จงทำการ plot ให้เหมือนกับรูปตัวอย่างด้านล่างนี้**

```python
sns.boxplot(x="class", y="age", data=titanic, palette='rainbow')
```

**จงทำการ plot ให้เหมือนกับรูปตัวอย่างด้านล่างนี้**

```python
sns.swarmplot(x="class", y="age", data=titanic, palette='rainbow')
```

**จงทำการ plot ให้เหมือนกับรูปตัวอย่างด้านล่างนี้**

```python
sns.countplot(x='sex', data=titanic)
```

**จงทำการ plot ให้เหมือนกับรูปตัวอย่างด้านล่างนี้**

```python
sns.heatmap(titanic.corr(),cmap='coolwarm')
```

**จงทำการ plot ให้เหมือนกับรูปตัวอย่างด้านล่างนี้**

```python
g = sns.FacetGrid(titanic, col="sex")
g.map(plt.hist, "age")
```
